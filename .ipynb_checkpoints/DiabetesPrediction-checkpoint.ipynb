### CO Relation Matrix

```python
correlation = data.corr ()
print (correlation)
```

* Pregnancies and Age (0.54):
There is a moderate positive correlation between the number of pregnancies and age. This suggests that as age increases, the number of pregnancies tends to increase.

* Glucose and Outcome (0.47):
There is a moderate positive correlation between glucose levels and the outcome of diabetes. Higher glucose levels are associated with a higher likelihood of a positive diabetes outcome.

* Blood Pressure and Age (0.24):
There is a weak positive correlation between blood pressure and age. Older individuals may tend to have slightly higher blood pressure.

* Skin Thickness and BMI (0.39):
There is a moderate positive correlation between skin thickness and BMI (Body Mass Index). Higher BMI is associated with thicker skin.

* Insulin and Glucose (0.33):
There is a moderate positive correlation between insulin and glucose levels. Higher glucose levels are associated with higher insulin levels.

* Pregnancies and Outcome (0.22):
There is a weak positive correlation between the number of pregnancies and the diabetes outcome. More pregnancies might be weakly associated with a higher likelihood of a positive diabetes outcome.

#### Visualizing the Correlation

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

#### Training the Model With the Health of Train Test Split

#### Train test split

```python
X = data.drop ('Outcome', axis = 1)
Y = data ['Outcome']
X_train, X_test, Y_train, Y_test= train_test_split (X,Y, test_size=0.2)
```

#### Training the model

```python
model= LogisticRegression ()
model.fit (X_train, Y_train)
```

#### Making Prediction

```python
prediction = model.predict (X_test)
```

```python
print (prediction)
```

```python
accuracy= accuracy_score (prediction, Y_test)
```

```python
print (accuracy)
```
